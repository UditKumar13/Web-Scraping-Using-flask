```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
import requests
##to fetch the webpage

from bs4 import BeautifulSoup
##to parse the webpage 
%matplotlib inline
```

```python
url='https://www.imdb.com/chart/top'
## fetch the web page
r=requests.get(url)
info=r.text
soup=BeautifulSoup(r.text,'html.parser')
TITLES=[]
table_inf=soup.find_all('tbody',{'class':'lister-list'})
doc=table_inf[0]
rows=doc.find_all('tr')
test_row=rows[0]
#first tr ka information 
```

```python

title_column=test_row.find('td',{'class':'titleColumn'})
movie_title=title_column.a.text
TITLES=[]
STORY=[]
DATE=[]
TAGLINE=[]
GENRE=[]
RELEASE_DATE=[]
BUDGET=[]
EARNING=[]


```

```python
count=0
movie_root_url = 'https://www.imdb.com'
for row in rows:
    if count>8:
        break
    count+=1
    title_column=row.find('td',{'class':'titleColumn'})
    #iterate to each row and collect data abt td and sotres it into title_column
    movie_title=title_column.a.text 
    p= movie_root_url + title_column.a['href']
    uk=requests.get(p)
    ##making a get request every time the loop runs , it will take us to next movie's url 
    inf0r=uk.text
    ##Using beautiful soup
    soup=BeautifulSoup(uk.text,'html.parser')
    page_inf=soup.find_all('div',{'class':'main','id':'main_bottom'})
    doc_1=page_inf[0]
    imp_doc=doc_1.find('div',{'class':'article','id':'titleStoryLine'})
    #THIS IMP HAS 3 ATTRIBUTES STORY,GENRE ,TAGLINE
    
    story_line=imp_doc.find('div',{'class':'inline canwrap'})
    story=story_line.span.text
    tag_line=imp_doc.find('div',{'class':'txt-block'})
    tagline=tag_line.text.split(':')[1].strip()
    genre=tag_line=imp_doc.find('div',{'itemprop':'genre'}).a.text
    budget_bro=doc_1.find('div',{'id':'titleDetails'})
    bud=budget_bro.find_all('div',{'class':'txt-block'})
    
    budget=bud[6].text.split(':')[1].split('(')[0].strip()


    
    date=bud[3].text.split(':')[1].split('(')[0].strip()
     
    STORY.append(story)
    TITLES.append(movie_title)
    TAGLINE.append(tagline)
    GENRE.append(genre)
    DATE.append(date)
    BUDGET.append(budget)
    

    
    
    

    
print (count)    


    
```

```python
uk_df=pd.DataFrame({'Title': TITLES, 'Story': STORY, 'Tagline': TAGLINE, 'Genre': GENRE, 'Release-date': DATE, 'Budget': BUDGET},
                     columns=['Title', 'Story', 'Tagline', 'Genre', 'Release-date'])
```
